```python
# Use to see what we can put on the graphs https://telemetry.sem-app.com/wiki/doku.php/telemetry_data/channel_descriptions
# This one is for use with BE cars. This is becuase some of the headers are spesific to an eletronic system and will not work with a ICE car

import pandas as pd
import plotly.io as pio
import plotly.graph_objects as go
import numpy as np

pd.options.plotting.backend = 'plotly'
pio.templates.default = 'plotly_dark'

#Replace this file to change data
df = pd.read_csv('./Data/Sample/BE_sample_data.csv', sep=',', low_memory=False)

# df.columns = [col.lower() for col in df.columns]
# df.reset_index(drop=True, inplace=True)
# df.ffill(inplace=True)
# df.head(10)
```

```python
# this shows us the map of the track. 
df.plot(x='gps_longitude', y='gps_latitude', color='lap_lap')
```

```python
# Plot visulizes the Instantaneous current measured by Joule Meter at diffrent points on the track
# Gives a good idea at where the speed changes where on the track
df.plot(x='gps_longitude', y='gps_latitude', color='jm3_current', kind='scatter', facet_col='lap_lap', facet_col_wrap=4)
```

```python
# This is the ammount of Instantaneous current measured by Joule Meter at a distance.
# Shows us how entinse and long the burns where
df.plot(x='lap_dist', y='jm3_current', color='lap_lap')
```

```python
# This is the total ammount of enerfy used per lap as compared to distance. (Lowest is best)
# Used to determine wich lap was best
df.plot(x='lap_dist', y='lap_jm3_netjoule', color='lap_lap')
```

```python
# Speed at diffrent points of the lap
df.plot(x='lap_dist', y='gps_speed', color='lap_lap')
```
